## Moving average method
we could see that the data is not stationary. The basic assumption before applying stochastic models is that the time series should be stationary.So, our first goal is to make the time series into stationary.Next step is to find seasonality (s). we plot the moving average to show the relation(is it stationary or not).
you must all be wondering what basically is stochastic model. it's a model which has bit of randomness in it i.e the output can't be determined as incase of deterministic model

## Moving Average(4 months)

```python
four_months_moving_average = indexedData.rolling(window=4).mean()
plt.plot(indexedData, color='blue', label='Original')
plt.plot(four_months_moving_average, color='red', label='Rolling Mean')
plt.legend(loc='best')
plt.title('4 Months Moving Average')
```

## Dickey Fuller Test(another method)
It tests the null hypothesis that a unit root present in autoregressive model.the values that we obtain after calling adfuller function are Test Statistic, p-value, No. of Lags used, Number of observations used and critical value.if the p-value is very less than we can say it's sationary else it's not.

https://en.wikipedia.org/wiki/Unit_root_test

https://en.wikipedia.org/wiki/Dickey%E2%80%93Fuller_test

```python
dftest = adfuller(indexedData['#Passengers'], autolag='AIC')
```

```python
dftest
```

```python
dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', 'No. of Lags used', 'Number of observations used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)' %key] = value
```

```python
dfoutput
```

```python
indexedData_logScale= np.log(indexedData)

plt.plot(indexedData_logScale, 'blue')

## we just made the value under the log scale
##there is no variation in the plot
```

```python
movingAverage = indexedData_logScale.rolling(window=12).mean()
movingSTD = indexedData_logScale.rolling(window=12).std()

plt.plot(indexedData_logScale, color='blue')
plt.plot(movingAverage, color='red')
##plot of std & mean
```

```python
##different transformation methods used

dataLogScaleMinusMovingAverage = indexedData_logScale - movingAverage
dataLogScaleMinusMovingAverage.dropna(inplace=True)
dataLogScaleMinusMovingAverage.head()


```

```python
plt.plot(dataLogScaleMinusMovingAverage)
##its stationary
```

```python
decomposition = seasonal_decompose(indexedData_logScale)
```

```python
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid
```

```python
fig = decomposition.plot()
fig.set_figwidth(12)
fig.set_figheight(8)
plt.suptitle('Decomposition of multiplicative time series')
```

```python
decomposedLogData = residual
decomposedLogData.dropna(inplace=True)
```

```python
##While fitting an arima model, we need to find correct 'p', 'd' and 'q'.
##We find 'd' by differencing the data number of times till it becomes stationary.

##this plots are superuseful while determining the P and Q value


fig, axes = plt.subplots(1, 2, sharey=False, sharex=False)
fig.set_figwidth(12)
fig.set_figheight(4)
plot_acf(dataLogScaleMinusMovingAverage, lags=20, ax=axes[0], alpha=0.5)
plot_pacf(dataLogScaleMinusMovingAverage, lags=20, ax=axes[1], alpha=0.5)
plt.tight_layout()
```

#### ARIMA model(p=2,q=1,d=2) . simialry done below.

```python
model = ARIMA(indexedData_logScale, order=(2,1,2))
results_ARIMA = model.fit(disp=-1)
plt.plot(dataLogScaleMinusMovingAverage, color='blue')
plt.plot(results_ARIMA.fittedvalues, color='red')
plt.title('RSS: %4f'% sum((results_ARIMA.fittedvalues - dataLogScaleMinusMovingAverage['#Passengers'])**2))
print('Plotting ARIMA model')
```

```python
model = ARIMA(indexedData_logScale, order=(2,1,0))
results_AR = model.fit()
plt.plot(dataLogScaleMinusMovingAverage)
plt.plot(results_AR.fittedvalues, color='red')
plt.title('RSS: %4f'% sum((results_AR.fittedvalues - dataLogScaleMinusMovingAverage['#Passengers'])**2))
print('Plotting ARIMA model')
```

```python
predictions_ARIMA_diff = pd.Series(results_ARIMA.fittedvalues, copy=True)
predictions_ARIMA_diff.head()
```

```python
plt.plot(predictions_ARIMA_diff)
```

convert the differencing to log scale is to add these differences consecutively to the base number.

```python
predictions_ARIMA_diff_cumsum = predictions_ARIMA_diff.cumsum()
predictions_ARIMA_diff_cumsum.head()
```

```python
predictions_ARIMA_log = pd.Series(indexedData_logScale['#Passengers'].ix[0], index=indexedData_logScale.index)
predictions_ARIMA_log = predictions_ARIMA_log.add(predictions_ARIMA_diff_cumsum, fill_value=0)
predictions_ARIMA_log.head()
```

```python
predictions_ARIMA = np.exp(predictions_ARIMA_log)
plt.plot(indexedData)
plt.plot(predictions_ARIMA)
```
